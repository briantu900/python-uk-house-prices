```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# **Part 1**

```python
df = pd.read_csv('UK_house_price_2023.csv')
df.info()

df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%Y %b')

df.describe()
```

```python
df.head()
```

```python
df.boxplot("12 month percentage change ")
#outliers shown as circles
```

```python

figure1 = px.line(df, x = 'date', y = '12 month percentage change ', title="% Change in UK House Prices 2006-2023",)
figure1.update_xaxes(title='Date')
figure1.update_yaxes(title='% Change')

figure1.show()

#significant drop starting at 2007 caused by 2007-2008 financial crash
#house prices rise again due to recovering economy
#unstable prices during covid

#df['date'] = pd.to_datetime(df['date'])
#df['date'] = df['date'].dt.strftime('%Y')
#df.plot.line(x = 'date')
```

```python

date_range = df[(df['date'] >= '2020') & (df['date'] <= '2023')]

figure2 = px.line(date_range, x = 'date', y = '12 month percentage change ', title="% Change in UK House Prices 2020-2023",)
figure2.update_xaxes(title='Date')
figure2.update_yaxes(title='% Change')

figure2.show()

#prices during covid

#date_range = df[(df['date'] >= '2020') & (df['date'] <= '2023')]
#date_range.plot.line(x = 'date')
```
